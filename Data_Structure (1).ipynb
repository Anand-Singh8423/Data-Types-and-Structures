#Data Types and Structures Questions
**Question-1.**What are data structures, and why are they important?

**Answer-1.**Data structures organize and store data efficiently for easy access and modification. They improve performance, clarity, and resource management in programs.

**Question-2.**Mutable vs Immutable (with examples):

Answer-Mutable: Can be changed. e.g., List → my_list = [1, 2]; my_list[0] = 10

Immutable: Cannot be changed. e.g., String → s = "hello"; s[0] = "H" → Error

**Question-3.**List vs Tuple:

List: Mutable, slower, uses []

Tuple: Immutable, faster, uses ()
Example: a = [1, 2], b = (1, 2)

**Question-4.**How do dictionaries store data?
As key-value pairs in a hash table, allowing fast access via keys.

**Question-5.**Why use a set instead of a list?
To ensure uniqueness and perform fast union/intersection operations.

**Question-6.**String vs List:

String: Immutable, only characters

List: Mutable, can contain any data type

**Question-7.**How do tuples ensure data integrity?
Their immutability prevents accidental changes.

**Question-8.**What is a hash table?
A structure that maps keys to values using a hash function. Python dictionaries are built on hash tables.

**Question-9.**Can lists contain different data types?
Yes. Example: [1, "two", 3.0, [4, 5]]

**Question-10.**Why are strings immutable in Python?
For performance, safety (e.g., in multithreading), and use as dictionary keys.

**Question-11.**Advantages of dictionaries over lists:

Fast lookup

Descriptive keys

Dynamic and flexible

Prevents duplicate keys

**Question-12.**When to prefer tuples over lists:
When data shouldn't change—like storing fixed coordinates.

**Question-13.**How do sets handle duplicates?
Sets ignore duplicates. Example: {1, 2, 2} → {1, 2}

**Question-14.**"in" keyword in list vs dictionary:

List: Checks values

Dictionary: Checks keys only

**Question-15.**Can you modify tuple elements?
No. Tuples are immutable—any attempt causes an error.

**Question-19.**Why are dictionaries unordered (until Python 3.7)?
Dictionaries used to store elements based on hash, not insertion order.
